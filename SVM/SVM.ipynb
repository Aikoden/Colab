```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import normalize
from collections import Counter

# 数据集列表
datasets = [
    'yXT_yeast.csv',
    'yXT_winequality-white.csv',
    'yXT_wine.csv',
    'yXT_seeds.csv',
    'yXT_car.csv',
    'yeast_out.csv'
]

# 遍历每个数据集
for dataset in datasets:
    # 读取数据集
    data = pd.read_csv(dataset, header=None)

    # 分离特征和标签
    X = data.iloc[:, 1:]  # 特征列
    y = data.iloc[:, 0]   # 标签列

    # 计算每个类别的样本数量
    class_counts = Counter(y)

    # 确定正类（样本数量最多的类别）
    positive_class = class_counts.most_common(1)[0][0]

    # 转换标签
    y_binary = y.apply(lambda x: 1 if x == positive_class else -1)

    # 划分训练集和测试集
    X_train, X_test, y_train, y_test = train_test_split(
        X, y_binary, test_size=0.3, random_state=42
    )

    # 定义归一化方法
    normalizations = {
        'L1 Normalization': normalize(X_train, norm='l1'),
        'L2 Normalization': normalize(X_train, norm='l2')
    }


    # 遍历每种归一化方法
    for norm_name, X_train_norm in normalizations.items():
        # 对测试集进行相同的归一化处理
        norm = norm_name.split()[0].lower()  # 获取 'l1' 或 'l2'
        X_test_norm = normalize(X_test, norm=norm)

        # 计算正则化参数 C = 1 / n
        n = len(y_train)  # 训练样本的数量
        C = n

        # 初始化线性SVM分类器
        model = SVC(kernel='linear', C=C, random_state=42)

        # 训练模型
        model.fit(X_train_norm, y_train)

        # 预测
        y_pred = model.predict(X_test_norm)

        # 评估模型
        accuracy = accuracy_score(y_test, y_pred)

        # 打印结果
        print(f"({dataset}) use {norm_name} SVM:")
        print(f"accuracy: {accuracy:.2f}\n")




```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import normalize
import pandas as pd
import numpy as np

datasets = [
    "yXT_yeast.csv",
    "yXT_wine.csv",
    "yXT_seeds.csv",
    "yXT_car.csv",
    "yeast_out.csv",
]
lambda_factors = [0.1, 1, 10]
for dataset in datasets:

    data = pd.read_csv(dataset, header=None)
    X = data.iloc[:, 1:]
    y = data.iloc[:, 0]
    y = y.astype(int)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    normalizations = {
        "Original": X_train,
        "L1 Normalization": normalize(X_train, norm="l1"),
        "L2 Normalization": normalize(X_train, norm="l2"),
    }

    for norm_name, X_train_norm in normalizations.items():
        X_test_norm = (
            normalize(X_test, norm=norm_name.split()[0].lower())
            if norm_name != "Original"
            else X_test
        )
        n = len(y_train)

        best_lambda = None
        best_score = -np.inf

        for lambda_factor in lambda_factors:
            lambda_value = lambda_factor / n
            C = 1 / lambda_value

            model = SVC(kernel="linear", C=C, decision_function_shape="ovr", random_state=42)

            kf = KFold(n_splits=5, shuffle=True, random_state=42)
            try:
                scores = cross_val_score(model, X_train_norm, y_train, cv=kf, scoring='accuracy')
                mean_score = np.mean(scores)
                if mean_score > best_score:
                    best_score = mean_score
                    best_lambda = lambda_value
                    best_lambda_str = f"{lambda_factor}/n"
            except ValueError as e:
                print(f"Error during cross-validation for λ={lambda_value}: {e}")

        C_best = 1 / best_lambda if best_lambda is not None else 1
        model_best = SVC(kernel="linear", C=C_best, decision_function_shape="ovr", random_state=42)
        model_best.fit(X_train_norm, y_train)
        y_pred = model_best.predict(X_test_norm)
        accuracy = accuracy_score(y_test, y_pred)

        print(f"Dataset: {dataset}, Normalization: {norm_name}")
        print(f"Best λ: {best_lambda_str}")
        print(f"Accuracy: {accuracy:.2f}\n")









```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import normalize
import pandas as pd
import numpy as np

# 数据集列表
datasets = [
    "yXT_yeast.csv",
    "yXT_wine.csv",
    "yXT_seeds.csv",
    "yXT_car.csv",
    # "yXT_winequality-white.csv"
]

# 正则化参数 λ 值列表
lambda_values = [1, 0.1, 0.01]

# 结果存储
results = {}

# 处理每个数据集
for dataset in datasets:
    print(f"Processing dataset: {dataset}")
    data = pd.read_csv(dataset, header=None)
    X = data.iloc[:, 1:]  # 特征
    y = data.iloc[:, 0]    # 标签
    y = y.astype(int)  # 确保标签为整数类型

    # 划分数据集为训练集和测试集
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

    # 定义三种正则化方法
    normalizations = {
        "Original": X_train,
        "L1 Normalization": normalize(X_train, norm="l1"),
        "L2 Normalization": normalize(X_train, norm="l2"),
    }

    # 存储当前数据集的结果
    results[dataset] = {}

    # 对每种正则化方法进行处理
    for norm_name, X_train_norm in normalizations.items():
        X_test_norm = (
            normalize(X_test, norm=norm_name.split()[0].lower())
            if norm_name != "Original"
            else X_test
        )

        best_lambda = None
        best_score = -np.inf

        # 对每个 λ 值进行交叉验证
        for lambda_factor in lambda_values:
            C = 1 / lambda_factor  # C 值是 λ 的倒数
            model = SVC(kernel="linear", C=C)

            try:
                scores = cross_val_score(model, X_train_norm, y_train, scoring='accuracy')
                mean_score = np.mean(scores)
                if mean_score > best_score:
                    best_score = mean_score
                    best_lambda = lambda_factor
            except ValueError as e:
                print(f"Error during cross-validation for λ={lambda_factor}: {e}")

        # 用最优 λ 值训练模型并计算测试集准确率
        C_best = 1 / best_lambda if best_lambda is not None else 1
        model_best = SVC(kernel="linear", C=C_best)
        try:
            model_best.fit(X_train_norm, y_train)
            y_pred = model_best.predict(X_test_norm)
            accuracy = accuracy_score(y_test, y_pred)
        except ValueError as e:
            print(f"Error during model training or prediction: {e}")
            accuracy = None

        # 存储结果
        results[dataset][norm_name] = {
            "best_lambda": best_lambda,
            "accuracy": accuracy
        }

        # 打印结果
        print(f"Normalization: {norm_name}")
        print(f"Best λ: {best_lambda if best_lambda is not None else 'Not determined'}")
        print(f"Accuracy: {accuracy:.2f}\n" if accuracy is not None else "Accuracy: Not available\n")

# 输出最终结果
for dataset, res in results.items():
    print(f"Final Results for {dataset}:")
    for norm_name, metrics in res.items():
        print(f"{norm_name}: Best λ = {metrics['best_lambda']}, Accuracy = {metrics['accuracy']:.2f}")
    print("\n")
```
