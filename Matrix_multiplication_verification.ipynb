```python
import numpy as np
import time

# 计算 A(B𝑥) 的时间和结果
def calculate_A_Bx(A, B, x):
    start_time = time.time()

    # 计算 Bx
    Bx = np.dot(B, x)

    # 计算 A(Bx)
    ABx = np.dot(A, Bx)

    end_time = time.time()
    elapsed_time = end_time - start_time

    return ABx, elapsed_time

# 直接计算 AB 的时间和结果
def calculate_AB(A, B):
    start_time = time.time()

    # 直接计算 AB
    AB = np.dot(A, B)

    end_time = time.time()
    elapsed_time = end_time - start_time

    return AB, elapsed_time

# 设置不同的矩阵维度
dimensions = [10, 50, 100, 200]

for N in dimensions:
    print(f"矩阵维度为 {N}x{N}:")

    # 生成随机矩阵 A 和 B
    A = np.random.randint(10, size=(N, N))
    B = np.random.randint(10, size=(N, N))

    # 随机生成向量 x
    x = np.random.randint(2, size=(N, 1))

    # 计算 A(B𝑥) 的时间和结果
    ABx, time_A_Bx = calculate_A_Bx(A, B, x)

    # 直接计算 AB 的时间和结果
    AB, time_AB = calculate_AB(A, B)

    # 输出结果
    print(f"A(B𝑥) 计算时间: {time_A_Bx:.6f} 秒")
    print(f"直接计算 AB 的时间: {time_AB:.6f} 秒")

    # 验证 A(B𝑥) 是否等于 AB
    if np.array_equal(ABx, AB):
        print("A(B𝑥) 和直接计算 AB 的结果相等")
    else:
        print("A(B𝑥) 和直接计算 AB 的结果不相等")

    print("-----------------------------------")
```

```python
import numpy as np
import time

# 定义已知的矩阵 A 和 B
A = np.array([
    [5, 0, 3, 3, 7],
    [9, 3, 5, 2, 4],
    [7, 6, 8, 8, 1],
    [6, 7, 7, 8, 1],
    [5, 9, 8, 9, 4]
])

B = np.array([
    [3, 0, 3, 5, 0],
    [2, 3, 8, 1, 3],
    [3, 3, 7, 0, 1],
    [9, 9, 0, 4, 7],
    [3, 2, 7, 2, 0]
])

N = 5  # 矩阵的维度

D = np.dot(A, B)  # 正确的结果矩阵

# 函数：在矩阵 D 中引入错误
def introduce_errors(D, num_errors):
    C = D.copy()
    n = D.shape[0]
    for _ in range(num_errors):
        i = np.random.randint(n)
        j = np.random.randint(n)
        C[i, j] += np.random.randint(1, 10)
    return C

# 引入错误
num_errors = 3
C = introduce_errors(D, num_errors)

# Freivalds算法的实现及其运行时间测量
def freivalds(A, B, C, k=10):
    n = A.shape[0]
    start_time = time.time()
    for _ in range(k):
        r = np.random.randint(3, size=(n, 1))  # 随机向量 r
        Br = np.dot(B, r)
        ABr = np.dot(A, Br)
        Cr = np.dot(C, r)
        if not np.array_equal(ABr, Cr):
            end_time = time.time()
            return False, end_time - start_time
    end_time = time.time()
    return True, end_time - start_time

# 计算矩阵乘积 A * B 并测量时间
def calculate_AB(A, B):
    start_time = time.time()
    AB = np.dot(A, B)
    end_time = time.time()
    elapsed_time = end_time - start_time
    return AB, elapsed_time

# 运行Freivalds算法并测量时间
result, freivalds_time = freivalds(A, B, C)
print(f"Freivalds算法验证结果: {result}")
print(f"Freivalds算法运行时间: {freivalds_time:.6f} 秒")

# 计算 A * B 的时间
AB, time_AB = calculate_AB(A, B)
print(f"矩阵乘法 A * B 运行时间: {time_AB:.6f} 秒")

# 计算Freivalds算法的错误概率
def calculate_error_probability(A, B, C, trials=1000):
    n = A.shape[0]
    error_count = 0
    for _ in range(trials):
        r = np.random.randint(3, size=(N, 1))
        Bx = np.dot(B, r)
        ABx = np.dot(A, Bx)
        Cx = np.dot(C, r)
        if not np.array_equal(ABx, Cx):
            error_count += 1
    return error_count / trials

# 计算错误概率
trials = 1000
error_probability = calculate_error_probability(A, B, C, trials)
print(f"错误概率: {error_probability:.4f}")

# 打印矩阵
print("矩阵 A:")
print(A)
print("矩阵 B:")
print(B)
print("矩阵 D (A * B 的结果):")
print(D)
print("矩阵 C (引入错误后的结果):")
print(C)
```

```python
import numpy as np

def freivalds_algorithm(A, B, C, k=5):
    n = A.shape[0]
    for _ in range(k):
        # ランダムベクトルを生成
        r = np.random.randint(0, 2, size=n)

        # (AB)rとCrを計算
        ABr = A @ (B @ r)
        Cr = C @ r
        # 等しいかチェック
        if not np.array_equal(ABr, Cr):
            return False  # 確実に等しくない

    return True  # おそらく等しい

# 使用例
n = 100
A = np.random.randint(low=0,high=100,size=(n,n))
B = np.random.randint(low=0,high=100,size=(n,n))
C = A @ B  # 正しい結果

print(freivalds_algorithm(A, B, C))  # Trueが出力されるはず

# エラーを導入
C[0, 0] += 1
print(freivalds_algorithm(A, B, C))  # Falseが出力されるはず
```

```python
import numpy as np

def freivalds_algorithm(A, B, C):
    n = A.shape[0]
    x = np.random.randint(0, 2, size=n)
    return np.array_equal(A @ (B @ x), C @ x)

def theoretical_probability(num_error_rows):
    return 1 - (1/2)**num_error_rows

def experimental_probability(A, B, C, num_trials=10000):
    count = 0
    for _ in range(num_trials):
        if not freivalds_algorithm(A, B, C):
            count += 1
    return count / num_trials

# 行列のサイズ
n = 100

# 行列の生成
A = np.random.randint(0, 100, size=(n, n))
B = np.random.randint(0, 100, size=(n, n))
C = A @ B

# エラーを導入する関数
def introduce_errors(C, num_error_rows):
    C_error = C.copy()
    error_rows = np.random.choice(n, num_error_rows, replace=False)
    for row in error_rows:
        C_error[row, np.random.randint(n)] ^= 1  # ビット反転でエラーを導入
    return C_error

# 異なるエラー行数でテスト
for num_error_rows in [1, 2, 3, 5, 10]:
    C_error = introduce_errors(C, num_error_rows)

    theo_prob = theoretical_probability(num_error_rows)
    exp_prob = experimental_probability(A, B, C_error)

    print(f"エラー行数: {num_error_rows}")
    print(f"理論的確率: {theo_prob:.4f}")
    print(f"実験的確率: {exp_prob:.4f}")
    print()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义 f5 函数
def f5(x):
    return np.sqrt(x[0] * x[1])

# 计算 f5 的梯度
def grad_f5(x):
    return np.array([
        0.5 * np.sqrt(x[1] / x[0]),
        0.5 * np.sqrt(x[0] / x[1])
    ])

# 计算 f5 的解析 Hessian 矩阵
def hessian_f5_analytical(x):
    x1, x2 = x
    sqrt_x1x2 = np.sqrt(x1 * x2)
    H = np.array([
        [-np.sqrt(x2) / (4 * x1**1.5), 1 / (4 * sqrt_x1x2)],
        [1 / (4 * sqrt_x1x2), -np.sqrt(x1) / (4 * x2**1.5)]
    ])
    return H

# 数值计算 Hessian 矩阵
def numerical_hessian(grad_f, x, epsilon=1e-5):
    n = len(x)
    hessian = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            x_plus_ei = x.copy()
            x_plus_ei[i] += epsilon
            grad_plus_ei = grad_f(x_plus_ei)

            x_minus_ei = x.copy()
            x_minus_ei[i] -= epsilon
            grad_minus_ei = grad_f(x_minus_ei)

            hessian[i, j] = (grad_plus_ei[j] - grad_minus_ei[j]) / (2 * epsilon)

    return hessian

# 主函数
def main(f, grad_f, true_hessian_func, func_name, x_min, x_max, y_min, y_max):
    x = np.linspace(x_min, x_max, 91)
    y = np.linspace(y_min, y_max, 91)

    points = np.array([[xi, yi] for xi in x for yi in y])

    frobenius_errors = []

    for point in points:
        if point[0] <= 0 or point[1] <= 0:
            continue
        true_hessian_matrix = true_hessian_func(point)
        approx_hessian_matrix = numerical_hessian(lambda x: grad_f(x), point)
        error = np.linalg.norm(true_hessian_matrix - approx_hessian_matrix, 'fro')
        frobenius_errors.append(error)

    frobenius_errors = np.array(frobenius_errors)
    sorted_errors = np.sort(frobenius_errors)[::-1]
    top_1_percent_index = max(int(len(sorted_errors) * 0.01), 1)
    top_1_percent_errors = sorted_errors[:top_1_percent_index]

    plt.figure(figsize=(10, 6))
    plt.plot(np.linspace(0, 0.01, len(top_1_percent_errors)), top_1_percent_errors, color='red', marker='o', linestyle='-')
    plt.xlabel("Point Index Percentage")
    plt.ylabel("Frobenius Error")
    plt.title(f"Top 1% Frobenius Errors for {func_name}")
    plt.xlim(0, 0.01)
    plt.grid(True)
    plt.yscale('log')

    # 保存图像
    filename = f"{func_name}_error_plot.png"
    plt.savefig(filename)
    print(f"Plot saved as {filename}")
    plt.show()

    # 打印结果
    print(f"Results for {func_name}:")
    print(f"Max Frobenius Error: {np.max(frobenius_errors):.2e}")
    print(f"Mean Frobenius Error: {np.mean(frobenius_errors):.2e}")
    print()

# 运行示例
main(f5, grad_f5, hessian_f5_analytical, "f5(x)", 0.1, 2, 0.1, 2)





```

```python
import numpy as np
import matplotlib.pyplot as plt
import jax.numpy as jnp
from jax import grad, hessian as jax_hessian

# 定义 f3 函数
def f3(x):
    x1, x2 = x
    return (1 - x1 * x2) * (1 + x1 * x2) / (x1 * x2)

# 计算 f3 的梯度
def grad_f3(x):
    x1, x2 = x
    term1 = (x1**2 + x2**2 - 1) / (x1**2 * x2**2)
    term2 = - (1 + x1**2 - x2**2) / (x1 * x2**2)
    return np.array([
        term1 - term2,
        term2 - (x1**2 + x2**2 - 1) / (x1**2 * x2)
    ])

# 计算 f3 的解析 Hessian 矩阵
def hessian_f3_analytical(x):
    x1, x2 = x
    H = np.array([
        [2 / (x1**3 * x2), 1 / (x1**2 * x2**2) - 1],
        [1 / (x1**2 * x2**2) - 1, 2 / (x1 * x2**3)]
    ])
    return H

# 使用 JAX 计算 Hessian 矩阵
def hessian_f3_jax(x):
    x_jax = jnp.array(x)
    return np.array(jax_hessian(lambda x: f3(x))(x_jax))

# 数值计算 Hessian 矩阵
def numerical_hessian(grad_f, x, epsilon=1e-5):
    n = len(x)
    hessian = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            x_plus_ei = x.copy()
            x_plus_ei[i] += epsilon
            grad_plus_ei = grad_f(x_plus_ei)

            x_minus_ei = x.copy()
            x_minus_ei[i] -= epsilon
            grad_minus_ei = grad_f(x_minus_ei)

            hessian[i, j] = (grad_plus_ei[j] - grad_minus_ei[j]) / (2 * epsilon)

    return hessian

# 主函数
def main(f, grad_f, true_hessian_func, jax_hessian_func, func_name, x_min, x_max, y_min, y_max):
    x = np.linspace(x_min, x_max, 41)
    y = np.linspace(y_min, y_max, 41)

    points = np.array([[xi, yi] for xi in x for yi in y])

    frobenius_errors = []

    for point in points:
        if point[0] <= 0 or point[1] <= 0:
            continue
        true_hessian_matrix = true_hessian_func(point)
        jax_hessian_matrix = jax_hessian_func(point)
        approx_hessian_matrix = numerical_hessian(lambda x: grad_f(x), point)
        error = np.linalg.norm(jax_hessian_matrix - approx_hessian_matrix, 'fro')
        frobenius_errors.append(error)

    frobenius_errors = np.array(frobenius_errors)
    sorted_errors = np.sort(frobenius_errors)[::-1]
    top_1_percent_index = max(int(len(sorted_errors) * 0.01), 1)
    top_1_percent_errors = sorted_errors[:top_1_percent_index]

    plt.figure(figsize=(10, 6))
    plt.plot(np.linspace(0, 0.01, len(top_1_percent_errors)), top_1_percent_errors, color='red', marker='o', linestyle='-')
    plt.xlabel("Point Index Percentage")
    plt.ylabel("Frobenius Error")
    plt.title(f"Top 1% Frobenius Errors for {func_name}")
    plt.xlim(0, 0.01)
    plt.grid(True)
    plt.yscale('log')

    plt.show()

    # 打印结果
    print(f"Results for {func_name}:")
    print(f"Max Frobenius Error: {np.max(frobenius_errors):.2e}")
    print(f"Mean Frobenius Error: {np.mean(frobenius_errors):.2e}")
    print()

# 运行示例
main(f3, grad_f3, hessian_f3_analytical, hessian_f3_jax, "f3(x)", 0.1, 2, 0.1, 2)


```

```python
import timeit

def bm(g,x,p):
    bin_x = bin(x)[2:]
    n = len(bin_x)  # Number of bits in the binary representation

    # Initialize A with g
    A = g

    # Process each bit of the exponent from high to low (starting from index 1)
    for i in range(1, n):
        # Square A
        A = (A * A) % p
        # If the current bit is 1, multiply A by g
        if bin_x[i] == '1':
            A = (A * g) % p

    return A

def powl(p, g, x):
    return pow(g, x, p)

def montgomery_ladder(p, g, x):
    r0, r1 = 1, g
    for bit in bin(x)[2:]:
        if bit == '0':
            r1 = (r0 * r1) % p
            r0 = (r0 * r0) % p
        else:
            r0 = (r0 * r1) % p
            r1 = (r1 * r1) % p
    return r0

def test_and_time(p, g, x):
    print(f"\nTesting with p={p}, g={g}, x={x}")

    bm_result = bm(g, x, p)
    pow_result = powl(p, g, x)
    montgomery_result = montgomery_ladder(p, g, x)

    print(f"Binary Method result: {bm_result}")
    print(f"Pow function result: {pow_result}")
    print(f"Montgomery Ladder result: {montgomery_result}")

    assert bm_result == pow_result == montgomery_result, "Results do not match!"

    time_bm = timeit.timeit('bm(p,g,x)', globals=globals(), number=1)
    time_pow = timeit.timeit('powl(p,g,x)', globals=globals(), number=1)
    time_montgomery = timeit.timeit('montgomery_ladder(p,g,x)', globals=globals(), number=1)

    print(f"time_bm={time_bm}")
    print(f"time_pow={time_pow}")
    print(f"time_montgomery={time_montgomery}")

# Test cases
test_cases = [
    (9887, 2, 2**10 - 1),  # Original test case
    (2307, 6, 6789),       # The case where the issue was noticed
    (101, 2, 27),          # Small numbers
    (2**31 - 1, 3, 1000000)  # Large prime modulus
]

for p, g, x in test_cases:
    test_and_time(p, g, x)
```
