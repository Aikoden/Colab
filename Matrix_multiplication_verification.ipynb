```python
import numpy as np

def freivalds_algorithm(A, B, C, k=5):
    n = A.shape[0]
    for _ in range(k):
        # „É©„É≥„ÉÄ„É†„Éô„ÇØ„Éà„É´„ÇíÁîüÊàê
        r = np.random.randint(0, 2, size=n)

        # (AB)r„Å®Cr„ÇíË®àÁÆó
        ABr = A @ (B @ r)
        Cr = C @ r
        # Á≠â„Åó„ÅÑ„Åã„ÉÅ„Çß„ÉÉ„ÇØ
        if not np.array_equal(ABr, Cr):
            return False  # Á¢∫ÂÆü„Å´Á≠â„Åó„Åè„Å™„ÅÑ

    return True  # „Åä„Åù„Çâ„ÅèÁ≠â„Åó„ÅÑ

# ‰ΩøÁî®‰æã
n = 100
A = np.random.randint(low=0,high=100,size=(n,n))
B = np.random.randint(low=0,high=100,size=(n,n))
C = A @ B  # Ê≠£„Åó„ÅÑÁµêÊûú

print(freivalds_algorithm(A, B, C))  # True„ÅåÂá∫Âäõ„Åï„Çå„Çã„ÅØ„Åö

# „Ç®„É©„Éº„ÇíÂ∞éÂÖ•
C[0, 0] += 1
print(freivalds_algorithm(A, B, C))  # False„ÅåÂá∫Âäõ„Åï„Çå„Çã„ÅØ„Åö
```

```python
import numpy as np

def freivalds_algorithm(A, B, C):
    n = A.shape[0]
    x = np.random.randint(0, 2, size=n)
    return np.array_equal(A @ (B @ x), C @ x)

def theoretical_probability(num_error_rows):
    return 1 - (1/2)**num_error_rows

def experimental_probability(A, B, C, num_trials=10000):
    count = 0
    for _ in range(num_trials):
        if not freivalds_algorithm(A, B, C):
            count += 1
    return count / num_trials

# Ë°åÂàó„ÅÆ„Çµ„Ç§„Ç∫
n = 100

# Ë°åÂàó„ÅÆÁîüÊàê
A = np.random.randint(0, 100, size=(n, n))
B = np.random.randint(0, 100, size=(n, n))
C = A @ B

# „Ç®„É©„Éº„ÇíÂ∞éÂÖ•„Åô„ÇãÈñ¢Êï∞
def introduce_errors(C, num_error_rows):
    C_error = C.copy()
    error_rows = np.random.choice(n, num_error_rows, replace=False)
    for row in error_rows:
        C_error[row, np.random.randint(n)] ^= 1  # „Éì„ÉÉ„ÉàÂèçËª¢„Åß„Ç®„É©„Éº„ÇíÂ∞éÂÖ•
    return C_error

# Áï∞„Å™„Çã„Ç®„É©„ÉºË°åÊï∞„Åß„ÉÜ„Çπ„Éà
for num_error_rows in [1, 2, 3, 5, 10]:
    C_error = introduce_errors(C, num_error_rows)

    theo_prob = theoretical_probability(num_error_rows)
    exp_prob = experimental_probability(A, B, C_error)

    print(f"„Ç®„É©„ÉºË°åÊï∞: {num_error_rows}")
    print(f"ÁêÜË´ñÁöÑÁ¢∫Áéá: {theo_prob:.4f}")
    print(f"ÂÆüÈ®ìÁöÑÁ¢∫Áéá: {exp_prob:.4f}")
    print()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ÂÆö‰πâ f5 ÂáΩÊï∞
def f5(x):
    return np.sqrt(x[0] * x[1])

# ËÆ°ÁÆó f5 ÁöÑÊ¢ØÂ∫¶
def grad_f5(x):
    return np.array([
        0.5 * np.sqrt(x[1] / x[0]),
        0.5 * np.sqrt(x[0] / x[1])
    ])

# ËÆ°ÁÆó f5 ÁöÑËß£Êûê Hessian Áü©Èòµ
def hessian_f5_analytical(x):
    x1, x2 = x
    sqrt_x1x2 = np.sqrt(x1 * x2)
    H = np.array([
        [-np.sqrt(x2) / (4 * x1**1.5), 1 / (4 * sqrt_x1x2)],
        [1 / (4 * sqrt_x1x2), -np.sqrt(x1) / (4 * x2**1.5)]
    ])
    return H

# Êï∞ÂÄºËÆ°ÁÆó Hessian Áü©Èòµ
def numerical_hessian(grad_f, x, epsilon=1e-5):
    n = len(x)
    hessian = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            x_plus_ei = x.copy()
            x_plus_ei[i] += epsilon
            grad_plus_ei = grad_f(x_plus_ei)

            x_minus_ei = x.copy()
            x_minus_ei[i] -= epsilon
            grad_minus_ei = grad_f(x_minus_ei)

            hessian[i, j] = (grad_plus_ei[j] - grad_minus_ei[j]) / (2 * epsilon)

    return hessian

# ‰∏ªÂáΩÊï∞
def main(f, grad_f, true_hessian_func, func_name, x_min, x_max, y_min, y_max):
    x = np.linspace(x_min, x_max, 91)
    y = np.linspace(y_min, y_max, 91)

    points = np.array([[xi, yi] for xi in x for yi in y])

    frobenius_errors = []

    for point in points:
        if point[0] <= 0 or point[1] <= 0:
            continue
        true_hessian_matrix = true_hessian_func(point)
        approx_hessian_matrix = numerical_hessian(lambda x: grad_f(x), point)
        error = np.linalg.norm(true_hessian_matrix - approx_hessian_matrix, 'fro')
        frobenius_errors.append(error)

    frobenius_errors = np.array(frobenius_errors)
    sorted_errors = np.sort(frobenius_errors)[::-1]
    top_1_percent_index = max(int(len(sorted_errors) * 0.01), 1)
    top_1_percent_errors = sorted_errors[:top_1_percent_index]

    plt.figure(figsize=(10, 6))
    plt.plot(np.linspace(0, 0.01, len(top_1_percent_errors)), top_1_percent_errors, color='red', marker='o', linestyle='-')
    plt.xlabel("Point Index Percentage")
    plt.ylabel("Frobenius Error")
    plt.title(f"Top 1% Frobenius Errors for {func_name}")
    plt.xlim(0, 0.01)
    plt.grid(True)
    plt.yscale('log')

    # ‰øùÂ≠òÂõæÂÉè
    filename = f"{func_name}_error_plot.png"
    plt.savefig(filename)
    print(f"Plot saved as {filename}")
    plt.show()

    # ÊâìÂç∞ÁªìÊûú
    print(f"Results for {func_name}:")
    print(f"Max Frobenius Error: {np.max(frobenius_errors):.2e}")
    print(f"Mean Frobenius Error: {np.mean(frobenius_errors):.2e}")
    print()

# ËøêË°åÁ§∫‰æã
main(f5, grad_f5, hessian_f5_analytical, "f5(x)", 0.1, 2, 0.1, 2)





```

```python
import numpy as np
import matplotlib.pyplot as plt
import jax.numpy as jnp
from jax import grad, hessian as jax_hessian

# ÂÆö‰πâ f3 ÂáΩÊï∞
def f3(x):
    x1, x2 = x
    return (1 - x1 * x2) * (1 + x1 * x2) / (x1 * x2)

# ËÆ°ÁÆó f3 ÁöÑÊ¢ØÂ∫¶
def grad_f3(x):
    x1, x2 = x
    term1 = (x1**2 + x2**2 - 1) / (x1**2 * x2**2)
    term2 = - (1 + x1**2 - x2**2) / (x1 * x2**2)
    return np.array([
        term1 - term2,
        term2 - (x1**2 + x2**2 - 1) / (x1**2 * x2)
    ])

# ËÆ°ÁÆó f3 ÁöÑËß£Êûê Hessian Áü©Èòµ
def hessian_f3_analytical(x):
    x1, x2 = x
    H = np.array([
        [2 / (x1**3 * x2), 1 / (x1**2 * x2**2) - 1],
        [1 / (x1**2 * x2**2) - 1, 2 / (x1 * x2**3)]
    ])
    return H

# ‰ΩøÁî® JAX ËÆ°ÁÆó Hessian Áü©Èòµ
def hessian_f3_jax(x):
    x_jax = jnp.array(x)
    return np.array(jax_hessian(lambda x: f3(x))(x_jax))

# Êï∞ÂÄºËÆ°ÁÆó Hessian Áü©Èòµ
def numerical_hessian(grad_f, x, epsilon=1e-5):
    n = len(x)
    hessian = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            x_plus_ei = x.copy()
            x_plus_ei[i] += epsilon
            grad_plus_ei = grad_f(x_plus_ei)

            x_minus_ei = x.copy()
            x_minus_ei[i] -= epsilon
            grad_minus_ei = grad_f(x_minus_ei)

            hessian[i, j] = (grad_plus_ei[j] - grad_minus_ei[j]) / (2 * epsilon)

    return hessian

# ‰∏ªÂáΩÊï∞
def main(f, grad_f, true_hessian_func, jax_hessian_func, func_name, x_min, x_max, y_min, y_max):
    x = np.linspace(x_min, x_max, 41)
    y = np.linspace(y_min, y_max, 41)

    points = np.array([[xi, yi] for xi in x for yi in y])

    frobenius_errors = []

    for point in points:
        if point[0] <= 0 or point[1] <= 0:
            continue
        true_hessian_matrix = true_hessian_func(point)
        jax_hessian_matrix = jax_hessian_func(point)
        approx_hessian_matrix = numerical_hessian(lambda x: grad_f(x), point)
        error = np.linalg.norm(jax_hessian_matrix - approx_hessian_matrix, 'fro')
        frobenius_errors.append(error)

    frobenius_errors = np.array(frobenius_errors)
    sorted_errors = np.sort(frobenius_errors)[::-1]
    top_1_percent_index = max(int(len(sorted_errors) * 0.01), 1)
    top_1_percent_errors = sorted_errors[:top_1_percent_index]

    plt.figure(figsize=(10, 6))
    plt.plot(np.linspace(0, 0.01, len(top_1_percent_errors)), top_1_percent_errors, color='red', marker='o', linestyle='-')
    plt.xlabel("Point Index Percentage")
    plt.ylabel("Frobenius Error")
    plt.title(f"Top 1% Frobenius Errors for {func_name}")
    plt.xlim(0, 0.01)
    plt.grid(True)
    plt.yscale('log')

    plt.show()

    # ÊâìÂç∞ÁªìÊûú
    print(f"Results for {func_name}:")
    print(f"Max Frobenius Error: {np.max(frobenius_errors):.2e}")
    print(f"Mean Frobenius Error: {np.mean(frobenius_errors):.2e}")
    print()

# ËøêË°åÁ§∫‰æã
main(f3, grad_f3, hessian_f3_analytical, hessian_f3_jax, "f3(x)", 0.1, 2, 0.1, 2)


```
